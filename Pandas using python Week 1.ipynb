```python
import pandas as pd

data = pd.read_csv('Iris.csv')
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
data.drop('petal_width', axis=1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['sepal_size'] = data.sepal_length * data.sepal_width
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
plt.subplot(2, 2, 1)
sns.histplot(data['sepal_length'], kde=True)
plt.title('Sepal length of the species')
plt.subplot(2, 2, 2)
sns.histplot(data['sepal_width'], kde=True)
plt.title('Sepal width of the species')
plt.subplot(2, 2, 3)
sns.histplot(data['sepal_size'], kde=True)
plt.title('Dimensions of the species')
plt.tight_layout()
plt.show()
```
