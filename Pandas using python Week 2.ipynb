```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv("C:/Users/yathi/Downloads/Final_one_encoded_Energy_consumption.csv")
x = df [["Timestamp","Temperature","Humidity","SquareFootage","Occupancy","HVACUsage","LightingUsage","RenewableEnergy","DayOfWeek_Sunday", "DayOfWeek_Monday", "DayOfWeek_Tuesday","DayOfWeek_Wednesday","DayOfWeek_Thursday","DayOfWeek_Friday","DayOfWeek_Saturday","Holiday"]]
timestamp = x['Timestamp']
X_numeric = x.drop('Timestamp', axis=1)
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(X_numeric)
x_normalized_df = pd.DataFrame(x_normalized,columns=X_numeric.columns)
x_normalized_df['Timestamp'] = timestamp
print(x_normalized_df.head())
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df = pd.read_csv("C:/Users/yathi/Downloads/encoded_file_with_dayofweek_Energy_consumption.csv")
x = df [["Temperature","Humidity","SquareFootage","Occupancy","HVACUsage","LightingUsage","RenewableEnergy","DayOfWeek_Monday","DayOfWeek_Tuesday","DayOfWeek_Wednesday","DayOfWeek_Thursday","DayOfWeek_Friday","DayOfWeek_Saturday","DayOfWeek_Sunday","Holiday"]]
y = df [["EnergyConsumption"]]

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=42)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, Y_train.values.ravel())

importances = rf.feature_importances_

feature_importances = pd.DataFrame({
    'Feature': x.columns,
    'Importance': importances
})

feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

top_8_features = feature_importances.head(8)

print("Top 8 features based on importance:")
print(top_8_features)

selected_features = top_8_features['Feature'].tolist()
X_selected = x[selected_features]

print("\nSelected features DataFrame (X_selected):")
print(X_selected.head())
```
